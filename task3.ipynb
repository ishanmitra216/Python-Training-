```python
import numpy as np
matrix = np.random.randint(1, 50, (10, 10))

print("Original Matrix:\n", matrix)
add_matrix = matrix + 10
print("\nMatrix after Addition (+10):\n", add_matrix)
sub_matrix = matrix - 5
print("\nMatrix after Subtraction (-5):\n", sub_matrix)
mul_matrix = matrix * 2
print("\nMatrix after Multiplication (*2):\n", mul_matrix)
div_matrix = matrix / 2
print("\nMatrix after Division (/2):\n", div_matrix)
transpose_matrix = matrix.T
print("\nTranspose of Matrix:\n", transpose_matrix)
```

**NumPy** is a powerful Python library used for numerical computing.
In this program, a **10×10 matrix** is created using numpy.random.randint().
Various arithmetic operations such as addition, subtraction, multiplication, and division are performed on the matrix elements.
The transpose operation converts rows into columns. These operations are commonly used in data analytics, machine learning, and scientific computing.
