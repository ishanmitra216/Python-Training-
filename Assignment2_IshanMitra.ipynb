```python
def is_palindrome(num):
    original = num
    reverse = 0
    while num > 0:
        digit = num % 10
        reverse = reverse * 10 + digit
        num = num // 10
    if original == reverse:
        return True
    else:
        return False
# Example
print(is_palindrome(121))
```

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n-1)

print(factorial(5))
```

```python
num = int(input("Enter number: "))
temp = num
sum = 0

while temp > 0:
    digit = temp % 10
    sum += digit ** 3
    temp //= 10

if sum == num:
    print("Armstrong number")
else:
    print("Not Armstrong number")
```

```python
string = input("Enter string: ").lower()

vowels = 0
consonants = 0

for char in string:
    if char.isalpha():
        if char in "aeiou":
            vowels += 1
        else:
            consonants += 1

print("Vowels:", vowels)
print("Consonants:", consonants)
```

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
# find gcd
x, y = a, b
while y != 0:
    x, y = y, x % y
gcd = x
lcm = (a * b) // gcd
print("LCM is:", lcm)
```

```python
list1 = [1, 2, 2, 3, 4, 4, 5]

unique_list = []

for item in list1:
    if item not in unique_list:
        unique_list.append(item)

print("List without duplicates:", unique_list)
```

```python
num = int(input("Enter number: "))
flag = True

if num <= 1:
    flag = False
else:
    for i in range(2, num):
        if num % i == 0:
            flag = False
            break

if flag:
    print("Prime number")
else:
    print("Not prime number")
```
